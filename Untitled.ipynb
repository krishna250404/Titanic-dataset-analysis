```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Titanic.csv")
df[:50]
```

```python
df.info()
df.describe()
```

```python
df.nunique()
```

```python
sns.heatmap(df.corr(method="spearman", min_periods = 1, numeric_only = True))
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df
```

```python

df[["discard", "Fullname"]]=df["Name"].str.split(",", expand = True)
df.drop(columns=["Name", "discard"], inplace=True)
df
```

```python

translationtable = str.maketrans("","","()\"'")
df["Fullname"] = df["Fullname"].str.translate(translationtable)
df["Fullname"] = df["Fullname"].str.strip()
df
```

```python
df.drop(columns = ["Cabin"], inplace = True)
df
```

```python
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])
df
```

```python
df[['Ticket_prefix', 'Ticket_number']] = (df['Ticket'].str.extract(r'^(.*?)(\d+)$'))
df['Ticket_prefix'] = (df['Ticket_prefix'].str.strip().replace('', pd.NA))
df.drop(columns=["Ticket"], inplace=True)
df["Ticket_number"] = pd.to_numeric(df["Ticket_number"], errors='coerce').astype('Int64')
df
```

```python
df["Survived"] = df["Survived"].astype(bool)
df.info()
```

```python
df["Famsize"] = df["SibSp"] + df["Parch"] + 1
df
```
